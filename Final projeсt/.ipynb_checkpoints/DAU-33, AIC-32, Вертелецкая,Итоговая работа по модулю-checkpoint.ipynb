## Анализ данных для итоговой работы по модулю "Метрики, гипотезы точки роста" AIC-32

Исходные данные: база транзакций интернет-магазина (Великобритания).

Подгружаем данные и проверяем на качество и полноту.

```python
#Подключаем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Загружаем файл csv с данными
#К исходному xlsx файлу были добавлены колонки Quantity*UnitPrice (Сумма заказа = кол-во товара * цена за штуку), 	InvoceDateWithoutTime (Дата выставления счёта без учёта времени), InvoiceTime (Время выставления счёта)
df_start = pd.read_csv('Копия Data Sample (AIC) для итоговой работы (new) - Online Retail.csv', sep=',')
```

```python
#Смотрим основную информацию
df_start.head()
```

```python
#Пропуски в данных
pd.isna(df_start)
```

```python
#Пропуски в данных (кол-во)
pd.isna(df_start).sum()
```

```python
#Смотрим пример строки с пропущенными данными
df_start.iloc[204]
```

```python
#Замена пропусков на 0
df_start.fillna(0)
```

```python
df_start = df_start.fillna(0)
```

```python
#Смотрим пример строки с пропущенными данными после замены пропущенных значений на 0
df_start.iloc[204]
```

```python
#Неопределённые данные в описании. Пропуски были заменены на 0; ? и ?? были выявлены при работе с документом в xlsx
df_filter_anomal_description = df_start[((df_start['Description'] == 0) | (df_start['Description'] == '?')) | (df_start['Description'] == '??')]
df_filter_anomal_description
```

В данных с пропусками и аномалиями в столбце 'Description' имеют нулевую 0.0 цену и стоимость. Предполагаем, что данные транзакции не были проведены. Поэтому разобьём данный датасет на 2: с пропусками в описании ('Description' - df_filter_anomal_description) и без пропусков (df)

```python
#df без пропусков, аномалий в столбце 'Description'
df = df_start[((df_start['Description'] != 0) & (df_start['Description'] != '?')) & (df_start['Description'] != '??')]
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#Средние значения по данным с пропусками в описании
df_filter_anomal_description.mean(axis=0)
```

```python
#Диаграмма столбчатая по количеству товара в счёте (-Х возврат, Х продажа) среди данных с пропущенными значениями в описании
plt.figure(figsize=(10, 10))
plt.bar(df_filter_anomal_description['InvoiceNo'], df_filter_anomal_description['Quantity'])
```

```python
#Неопределённые данные в пользователях
df_filter_anomal_customer = df['CustomerID']. isin([0])
df[df_filter_anomal_customer]
```

```python
#Средние значения по данным с пропусками в пользователях
df[df_filter_anomal_customer].mean(axis=0)
```

```python
#Сводная таблица по аномальным пользователям: номер счёта и его общая стоимость по всем позициям с сортировкой по убыванию
pd.pivot_table(df[df_filter_anomal_customer], index='InvoiceNo', values='Quantity*UnitPrice', aggfunc='sum', margins=True).sort_values(by='Quantity*UnitPrice', ascending=False)
```

```python
#Сводная таблица по аномальным пользователям: страна и количество пользователей с сортировкой по убыванию:
pd.pivot_table(df[df_filter_anomal_customer], index='Country', values='CustomerID', aggfunc='count', margins=True).sort_values(by='CustomerID', ascending=False)
```

98% (13296) неидентифицированных пользователей приходится на United Kingdom.

```python
#перевели столбец со стоимостью в числовой тип и сохранили изменения
df.UnitPrice = pd.to_numeric(df.UnitPrice.replace(',', '.', regex=True), errors='coerce')
```

```python
#перевели столбец с суммой заказа в числовой тип и сохранили изменения
df['Quantity*UnitPrice'] = pd.to_numeric(df['Quantity*UnitPrice'].replace(',', '.', regex=True), errors='coerce')
```

```python
#перевели столбец с датой в тип дата и сохранили изменения
df.InvoceDateWithoutTime = pd.to_datetime(df.InvoceDateWithoutTime)
```
